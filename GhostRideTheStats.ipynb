## Predictions

Use machine learning algorithms to pick the winners of this weeks games.

#### This week's matchups: (Week 12)

```python
#matchups = [['KS','MK'], ['MS','WD'], ['EF','AS'], ['CW','JM'], ['JN','NG'],['RG','PC']] # Week 11
matchups = [['MS','EF'],['KS','NG'],['MK','PC'],['CW','WD'],['JM','RG'],['AS','JN']]
print matchups
```

### Simulate the matchups and plot the results

For each team, compute the distribution function based on earlier results. Then sample from this distribution 10,000 times. Compare these simulations with the opponent's simulations and figure out who wins more often.

```python

xgrid = np.linspace(0,200, 100)

sb.set_context(rc={"figure.figsize": (8,12),'font.size': 16})

for idx, game in enumerate(matchups):
    
    density1 = kde.gaussian_kde(pfre[manager_initials.index(game[0]),:])
    density2 = kde.gaussian_kde(pfre[manager_initials.index(game[1]),:])
    
    sample1 = density1.resample(10000)
    sample2 = density2.resample(10000)
    
    outcomes = sample1-sample2
    team1wins = filter(lambda x: x>0,outcomes[0])
    team2wins = filter(lambda x: x<0,outcomes[0])
    if len(team1wins) > len(team2wins):
        percentwin = len(team1wins)/100.
        xlab = "%s wins %.2f%% of the time" %(game[0],percentwin)
    else:
        percentwin = len(team2wins)/100.
        xlab = "%s wins %.2f%% of the time" %(game[1],percentwin)

    ax = plt.subplot(3,2,idx+1)
    ax.plot(xgrid, density1(xgrid), 'g-',label=game[0])
    ax.plot(xgrid, density2(xgrid), 'b-',label=game[1])
    plt.xlim((0,170))
    ax.set_yticklabels([])
    
    green_line = mlines.Line2D([], [], color='green', label=game[0])    
    blue_line = mlines.Line2D([], [], color='blue', label=game[1])
    plt.legend([green_line, blue_line],[game[0],game[1]],loc=2)
    plt.xlabel(xlab)
```
